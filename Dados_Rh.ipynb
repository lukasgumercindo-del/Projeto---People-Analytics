```python
import pandas as pd 
import numpy as np

np.random.seed(42)

n_funcionarios = 1000

ids= [f"EMP{i:04d}" for i in range (1, n_funcionarios +1 )]
idades = np.random.randint(22, 56, size=n_funcionarios)
generos = np.random.choice(["Masculino", "Feminino", "Não-binário"], size=n_funcionarios, p=[0.48, 0.49, 0.03])

depts_lista = ["Vendas", "Tecnologia", "Recursos Humanos", "Marketing", "Operações"]
departamentos = np.random.choice(depts_lista, size=n_funcionarios, p=[0.30, 0.25, 0.08, 0.12, 0.25])

niveis = np.random.choice(["Junior", "Pleno", "Senior"], size=n_funcionarios, p=[0.50, 0.35, 0.15])

salarios = []
for nivel in niveis: 
    if nivel == "Junior":
        salarios.append(round(np.random.uniform(3500, 5000),2))
    elif nivel == "Pleno":
        salarios.append(round(np.random.uniform(5500, 8500),2))
    else: 
        salarios.append(round(np.random.uniform(9000, 15000),2))
        
tempo_empresa_meses = np.random.randint(1, 72, size=n_funcionarios)
satisfacao = np.random.randint(1, 6, size=n_funcionarios)
performance = np.random.randint(1, 6, size=n_funcionarios)

status = []
for i in range(n_funcionarios):
    if satisfacao[i] <= 2:
        status.append(np.random.choice(["Ativo", "Inativo"], p=[0.40, 0.60]))
    else:
        status.append(np.random.choice(["Ativo", "Inativo"], p=[0.90, 0.10]))

df_rh = pd.DataFrame({
    "id_funcionario": ids,
    "idade": idades,
    "genero": generos,
    "departamento": departamentos,
    "nivel": niveis,
    "salario": salarios,
    "tempo_casa_meses": tempo_empresa_meses,
    "nota_satisfacao": satisfacao,
    "nota_performance": performance,
    "status": status
})

df_rh.to_csv("dados_brutos_rh.csv", index=False)
print("Base de dados de People Analytics gerada com sucesso! O arquivo 'dados_brutos_rh.csv' foi criado.")

df_rh.head()
```
